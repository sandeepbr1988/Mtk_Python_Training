```python
''' 6. Write a python function to print the sum of digits for the given data and print the execution time (in seconds).
        Data:
            n = 5
        Expected Output:
            Time to sum of 1 to 5 and required time to calculate is : (15, 0.0) '''

import time

def sum_of_numbers(n):
    start_time = time.time()
    total = sum(range(1, n + 1))
    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"Time to sum of 1 to {n} and required time to calculate is : ({total}, {elapsed_time})")

n = 5
sum_of_numbers(n)
            
```

```python
''' 7. Write a Python program to hash a word.
        Data:
            x = [0,1,2,3,0,1,2,0,0,2,2,4,5,5,0,1,2,6,2,3,0,1,0,2,0,2]
            word = Input the word be hashed: rfdf
        Expected Output:
            Enter the word to be hashed: culture
            The coded word is: C210220 '''

x = [0,1,2,3,0,1,2,0,0,2,2,4,5,5,0,1,2,6,2,3,0,1,0,2,0,2]
word = input("Enter the word to be hashed: ").lower()
coded_word = word.capitalize()[0]
for letter in word:
    if 'a' <= letter <= 'z':
        idx = ord(letter) - ord('a')
        coded_word += str(x[idx])
    else:
        coded_word += letter
print("The coded word is:", coded_word)

```

```python
''' 9. Write a Python program to calculate the time runs (difference between start and current time) of a program.
        Data:
            times(5)
            times(4)
        Expected Output:
            0
            1
            2
            3
            4
            0.003652000000101907
            0
            1
            2
            3
            0.0070380999998178595 '''
import time

def times(n):
    start = time.perf_counter()
    for i in range(n):
        print(i)
    elapsed = time.perf_counter() - start
    print(elapsed)

times(5)
times(4)
```

```python
''' 13. Write a Python program to create all possible permutations from a given collection of distinct numbers.
        Data:
            Collection:  [1, 2, 3]
        Expected Output:
            Collection of distinct numbers: [[3, 2, 1], [2, 3, 1], [2, 1, 3], [3, 1, 2], [1, 3, 2], [1, 2, 3]] '''

from itertools import permutations
collection = [1, 2, 3]
all_permutations = [list(p) for p in permutations(collection)]
print("Collection of distinct numbers:", all_permutations)
```

```python
''' 14. Write a Python program to find the number of notes (Sample of notes: 10, 20, 50, 100, 200 and 500).
        Data:
            880
            1000
        Expected Output:
            6
            2 '''
def min_notes(amount):
    denominations = [500, 200, 100, 50, 20, 10]
    count = 0
    for note in denominations:
        if amount >= note:
            num_notes = amount // note
            count += num_notes
            amount -= num_notes * note
    return count
amounts = [880, 1000]
for amt in amounts:
    print(min_notes(amt))
```

```python
''' 15. Write a Python function to find the number of divisors of a given integer is even or odd.
        Data:
            x = 15, x = 12, x = 9, x = 6, x = 3
        Expected Output:
            4
            6
            3
            4
            2 '''

def count_divisors(x):
    count = 0
    for i in range(1, x + 1):
        if x % i == 0:
            count += 1
    return count
test_values = [15, 12, 9, 6, 3]
for x in test_values:
    print(count_divisors(x))
```

```python
''' 16. Write a Python function to get the smallest number from a list.
        Data:
            smallest([1, 2, -8, 0])
        Expected Output:
            -8 '''
def smallest(numbers):
    smallest_num = numbers[0]
    for num in numbers[1:]:
        if num < smallest_num:
            smallest_num = num
    return smallest_num
print(smallest([1, 2, -8, 0]))
```

```python
''' 17. Write a Python function to get the largest number from a list.
        Data:
            largest([1,2,-8,0])
        Expected Output:
            2 '''
def largest(numbers):
    largest_num = numbers[0]
    for num in numbers[1:]:
        if num > largest_num:
            largest_num = num
    return largest_num
print(largest([1, 2, -8, 0]))
     
```

```python
''' 18. Write a Python function to multiply all the items in a list.
        Data:
            list = [1, 2, -8]
        Expected Output:
            -16 '''

def multiply_list(items):
    result = 1
    for item in items:
        result *= item
    return result
my_list = [1, 2, -8]
print(multiply_list(my_list))
```

```python
''' 20. Write a Python program to remove duplicates from the data given below.
        Data:
            a = [10,20,30,20,10,50,60,40,80,50,40]
        Expected Output:
            [10, 20, 30, 50, 60, 40, 80] '''
a = [10, 20, 30, 20, 10, 50, 60, 40, 80, 50, 40]
result = []
seen = set()
for item in a:
    if item not in seen:
        result.append(item)
        seen.add(item)
print(result)
```
